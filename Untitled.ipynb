```python
def tf_calculator(term, document):
    counter = 0.0
    tokens = nltk.word_tokenize(document)
    for token in tokens:
        if(term == token):
            counter += 1
    tf_value = 0
    if ((counter <= 0)):
        tf_value = 0
    else:
        tf_value = 1 + math.log10(counter)
    return tf_value
```

```python
def idf_calculator(term, sentences):
    if (idf[term] > 0):
        return idf[term]
    print("calculating")
    count_documents = 0;
    for line in sentences:
        tokens = nltk.word_tokenize(line)
        for token in tokens:
            if term == token:
                count_documents += 1
                break
    idf[term] = math.log10(totalDocuments/count_documents)
    return idf[term]
```

```python
# tfidf calculator
def tfidf_calculator(term, document, corpus):
    tf = tf_calculator(term, document)
    print(tf)
    idf = idf_calculator(term, corpus)
    print(idf)
    return tf * idf

tfidf_calculator("است", lines[:1], lines[:5])
```
