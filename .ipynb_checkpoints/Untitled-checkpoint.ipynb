```python
import io, nltk, sys, gensim
from gensim.models.word2vec import LineSentence
from gensim.models import Word2Vec
from gensim import models
```

```python
lines = [line.rstrip('\n') for line in open('Data/HAM-Train.txt', encoding="utf8")]
totalDocuments = len(lines)
```

```python
stop_words = [line.rstrip('\n') for line in open('Stop_words2.txt', encoding="utf8")]
```

```python
puredFile = open('puredInput.txt', 'w', encoding="utf8")
```

```python
for line in lines:
    delimiter = line.find('@@@@@@@@@@')+10
#     puredFile.write(line[:delimiter+1])
    line = line[delimiter:]
    tokens = nltk.word_tokenize(line)
    for token in tokens:
        if token in stop_words:
            continue
        else:
            puredFile.write(token+" ")
    puredFile.write("\n")
puredFile.close()

lines = [line.rstrip('\n') for line in open('puredInput.txt', encoding="utf8")]
```

```python
# print(lines[:2])
sentences = LineSentence('puredInput.txt')
```

```python
model = Word2Vec(sentences, size=300, window=5, min_count=1, workers=4)
```

```python
model.wv['محصول']
```

```python
model.wv.most_similar('اداره')
```
